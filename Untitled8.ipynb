```python
livros = {}

def add():
    t = input("Título: ").title()
    if t in livros:
        print("Já cadastrado."); return
    livros[t] = {"autor": input("Autor: "), "Emprestar Livro": False}
    print("Adicionado!")

def emprestar():
    t = input("Título: ").title()
    if t not in livros:
        print("Não encontrado."); return
    if livros[t]["Emprestar Livro"]:
        print("Já emprestado."); return
    livros[t]["Emprestar Livro"] = True
    print("Emprestado!")

def devolver():
    t = input("Título: ").title()
    if t not in livros:
        print("Não encontrado."); return
    if not livros[t]["Emprestar Livro"]:
        print("Não estava emprestado."); return
    livros[t]["Emprestar Livro"] = False
    print("Devolvido!")

def listar():
    disp = [t for t, i in livros.items() if not i["Emprestar Livro"]]
    if disp:
        print("\nLivros disponíveis:")
        for t in disp: print(f"- {t} ({livros[t]['autor']})")
    else:
        print("Nenhum disponível.")

def buscar():
    t = input("Título: ").title()
    if t in livros:
        i = livros[t]
        s = "Emprestado" if i["Emprestar Livro"] else "Disponível"
        print(f"{t} - {i['autor']} ({s})")
    else:
        print("Não encontrado.")

while True:
    print("\n1.Adicionar Livro 2.Emprestar Livro 3.Devolver Livro 4.Listar Livros 5.Buscar Livro 0.Sair")
    o = input("Opção: ")
    if o == "1": add()
    elif o == "2": emprestar()
    elif o == "3": devolver()
    elif o == "4": listar()
    elif o == "5": buscar()
    elif o == "0": break
    else: print("Opção inválida.")4
```

```python
oficina = {"veiculos": {}, "servicos": []}

def reg_veiculo():
    p = input("Placa: ").upper()
    if p in oficina["veiculos"]:
        print("Veículo já cadastrado.")
        return
    dono = input("Dono: ")
    modelo = input("Modelo: ")
    oficina["veiculos"][p] = {"dono": dono, "modelo": modelo}
    print("Veículo adicionado!")

def reg_servico():
    p = input("Placa: ").upper()
    if p not in oficina["veiculos"]:
        print("Veículo não encontrado.")
        return
    s = input("Serviço: ")
    oficina["servicos"].append((p, s))
    print("Serviço registrado!")

def listar_veiculos():
    if not oficina["veiculos"]:
        print("Nenhum veículo cadastrado.")
    else:
        for p, v in oficina["veiculos"].items():
            print(f"{p} - {v['modelo']} ({v['dono']})")

def listar_servicos():
    if not oficina["servicos"]:
        print("Nenhum serviço registrado.")
    else:
        for p, s in oficina["servicos"]:
            print(f"{p} - {s}")

def buscar():
    p = input("Placa: ").upper()
    if p in oficina["veiculos"]:
        v = oficina["veiculos"][p]
        print(f"{p} - {v['modelo']} ({v['dono']})")
    else:
        print("Veículo não encontrado.")

while True:
    print("\n1.Reg. veículo 2.Reg. serviço 3.Listar veículos 4.Listar serviços 5.Buscar 0.Sair")
    o = input("Opção: ")
    if o == "1": reg_veiculo()
    elif o == "2": reg_servico()
    elif o == "3": listar_veiculos()
    elif o == "4": listar_servicos()
    elif o == "5": buscar()
    elif o == "0": break
    else: print("Opção inválida.")
```
