```python
#Parte Ricardo

#Cadastro de Alunos
lista =[]
listaAlunos = ["Lucas", "Maria", "João", "Pedro", "Joana"]

#Adicionar Aluno
listaAlunos.append("Paulo")
print(listaAlunos)

#Remover Aluno
listaAlunos.remove("Pedro")
print(listaAlunos)

#Buscar Aluno pelo nome

nome = input("Digite o nome do aluno que deseja buscar: ")

if nome in listaAlunos:
  print("Aluno encontrado!")
else:
    print("Aluno não encontrado!")

#Remoção do último item

listaAlunos = ["Lucas", "Maria", "João", "Pedro", "Joana"]
ultimo_aluno = listaAlunos.pop()
print(ultimo_aluno)
print(listaAlunos)

#Transformação dos Nomes da lista em maiúsculas

listaAlunos = ["Lucas", "Maria", "João", "Pedro", "Joana"]
for i in range(len(listaAlunos)):
  listaAlunos[i] = listaAlunos[i].upper()

print(listaAlunos)
#Verificação de texto com in

listaAlunos = ["Lucas", "Maria", "João", "Pedro", "Joana" ]
if "Lucas" in listaAlunos:
  print("Lucas está na lista!")

#Mostrar qual é o maior nome e qual o menor nome da lista

listaAlunos = ["Lucas", "Maria", "João", "Pedro", "Joana" ]
maior_nome = max(listaAlunos)
menor_nome = min(listaAlunos)
print("O maior nome da lista é:", maior_nome)
print("O menor nome da lista é:", menor_nome)
```

```python
#Controle de despesas pessoais do aluno

#Registrar valores de despesas até que o usuário digite "0"

despesas = []
while True:
  valor = float(input("Digite o valor da despesa (ou 0 para sair): "))
  if valor == 0:
    print("Saindo do programa de despesas.")
    break
  despesas.append(valor)

  #Calcular o total gasto

  total_gasto = sum(despesas)
  print("O total gasto até o momento é de R$ {:.2f}".format(total_gasto))

  #Calcular a média de gastos

  media_gastos = total_gasto / len(despesas)
  print("A média de gastos do sistema é de R$ {:.2f}".format(media_gastos))
```
